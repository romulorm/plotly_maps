```python
# imports
import pandas as pd
```

```python
vendas = pd.read_csv("vendas_por_cidades.csv")
display(vendas)
```

```python
import plotly.express as px

grafico = px.density_map(vendas, lon="geolocation_lng", lat="geolocation_lat", z="quantidade", zoom=3, radius=10, map_style="open-street-map")
grafico.update_layout(margin={"r":0, "t":0, "l":0, "b":0}) # Retira as margens do grÃ¡fico
grafico.show()
```
